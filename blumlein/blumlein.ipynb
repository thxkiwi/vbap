# Blumlein Configuration

- Constructed using analog circuits
- Transformers wired in a configuration to achieve sum and difference.
- Assmuing the original currents differ only in phase, 
- The current in the difference channel will be $\frac{\pi}{2}$ differnt in phase from the current in the summation channel.
- The difference current is then passed between two resistors *d* and *e* in series between which a capacitor (_"condensor"_) *f* forms a shunt arm.
- The voltage across the capacitor f will be in phase with that in the summation channel.
    - Passing the current in the summation channel is passed through a plain resistive attenuator network composed of series resistances *g*, *h*, and a shunt resistance *i*, a voltage is obtained which remains in phase with the voltage across the capacitor *f* in the difference channel.
- These two voltages are then combined and separated by a sum and difference process to produce two final channels.
    - The voltage in the first final channel will be the sum of these.
    - The voltage in the second final channel will be the difference of these.
    - Since these voltages were in phase the two final channels will be in phase and differ only in magnitude.
- By choosing
    - shunt resistance *i* in the summation channel
    - shunt capacitance *f* in the difference channel
    for a given frequency, any degree of amplitude difference in the final channels can be obtained for a given phase difference in the original channels.
- For low frequencies, for a given azimuth angle (_"obliquity of the source"_), phase difference between waves will vary proportionately with frequency, being very small for very low frequency. 
    - The higher the frequency, the current in the difference channel will be increasingly great compared with that in the summation channel.
    - The shunt capacitor in the difference circuit will have the effect of producing a fixed intensity difference in the final channels for a given azimuth angle at all low frequencies.
- For higher frequencies, it is not necessary to convert phase shifts into amplitude differences,
    - Simply reproduce amplitude differences.
    - The shunt capacitor in the difference circuit is therefore built out with a resistance *K* whose value is substantially equal to that of resistanece *i* (the shunt resistance in the summation channel).

- The capacitance of *f* is such that its impedance is small compared with that of the series resistances d and e over the whole working range,
- The resistance *K* is such that it equals the reactance of the capacitor at approximately the frequnecy above which it is desired not to convert phase differences into amplitude differences.
    - *K* is generally equal to *i*
        - amplitude differences for high frequencies are passed on unchanged.
- *f*, *d*, *e*, *K* are all chosen to minimize any amplitude differences that would be created by any phase differences.
    - recalling that amplitude/phase differences in one space (L/R) are phase/amplitude differences in the other space (M/S).
- The amount of modification to the summation and difference channels depends 
    - relative spacing of the microphones and of the loudspeakers,
    - size and position of the perceived sound stage (_"screen"_, recalling that Blumlein's system was originally designed to produce a sound perceived to have originated at certain distance across an image projected onto a screen).
    - For low frequencies w the degree of modification required in the difference channel as compared with the modification in the summation channel is given by :-

\begin{equation*}
K = \frac{2v}{jw} \cdot \frac{y}{\theta k} \cdot \frac{s}{x}
\end{equation*}

where,
- v = velocity of sound
    - approximately $343ms^{-1}$,
- y = fraction of half picture film width which the image of sound source is off center,
- $\theta$ = azimuth angle between the actual source and the vertical plane between the microphones,
    - $\theta$ is small
- k = effective distance apart of the microphones,
    - $k < \frac{1}{w}$
- s = width of the theater screen,
- x = distance apart of the loudspeakers.

\begin{equation*}
1 < \frac{s}{x} \le 1.3
\end{equation*}

and 

\begin{equation*}
s, x \ll \text{distance of the listener from the screen and loud speakers}
\end{equation*}
